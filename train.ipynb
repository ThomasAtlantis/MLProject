```python
import os
os.environ['CUDA_VISIBLE_DEVICES'] = '1'
import tensorflow, platform
print(f"{platform.version()}, Python {platform.python_version()}, TensorFlow {tensorflow.__version__}")
print(tensorflow.test.is_gpu_available())
```

```python
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D, BatchNormalization
from tensorflow.keras import optimizers
from tensorflow.keras import regularizers
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
x_train = x_train.astype('float32') / 255.0
y_train = to_categorical(y_train, 10)
x_test = x_test.astype('float32') / 255.0
y_test = to_categorical(y_test, 10)
```

```python
from numpy.random import RandomState
rdm = RandomState(1)
index_train = rdm.choice(a=x_train.shape[0], size=x_train.shape[0] // 2, replace=False)
index_test = rdm.choice(a=x_test.shape[0], size=x_test.shape[0] // 2, replace=False)
```

```python
weight_decay, nb_epoch, batch_size = 0.0005, 80, 32
```

```python
#layer1 32*32*3
model = Sequential()
model.add(Conv2D(64, (3, 3), padding='same',
input_shape=(32,32,3),kernel_regularizer=regularizers.l2(weight_decay)))
model.add(Activation('relu'))
model.add(BatchNormalization())
model.add(Dropout(0.3))
#layer2 32*32*64
model.add(Conv2D(64, (3, 3), padding='same',kernel_regularizer=regularizers.l2(weight_decay)))
model.add(Activation('relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))
#layer3 16*16*64
model.add(Conv2D(128, (3, 3), padding='same',kernel_regularizer=regularizers.l2(weight_decay)))
model.add(Activation('relu'))
model.add(BatchNormalization())
model.add(Dropout(0.4))
#layer4 16*16*128
model.add(Conv2D(128, (3, 3), padding='same',kernel_regularizer=regularizers.l2(weight_decay)))
model.add(Activation('relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))
#layer5 8*8*128
model.add(Conv2D(256, (3, 3), padding='same',kernel_regularizer=regularizers.l2(weight_decay)))
model.add(Activation('relu'))
model.add(BatchNormalization())
model.add(Dropout(0.4))
#layer6 8*8*256
model.add(Conv2D(256, (3, 3), padding='same',kernel_regularizer=regularizers.l2(weight_decay)))
model.add(Activation('relu'))
model.add(BatchNormalization())
model.add(Dropout(0.4))
#layer7 8*8*256
model.add(Conv2D(256, (3, 3), padding='same',kernel_regularizer=regularizers.l2(weight_decay)))
model.add(Activation('relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))
#layer8 4*4*256
model.add(Conv2D(512, (3, 3), padding='same',kernel_regularizer=regularizers.l2(weight_decay)))
model.add(Activation('relu'))
model.add(BatchNormalization())
model.add(Dropout(0.4))
#layer9 4*4*512
model.add(Conv2D(512, (3, 3), padding='same',kernel_regularizer=regularizers.l2(weight_decay)))
model.add(Activation('relu'))
model.add(BatchNormalization())
model.add(Dropout(0.4))
#layer10 4*4*512
model.add(Conv2D(512, (3, 3), padding='same',kernel_regularizer=regularizers.l2(weight_decay)))
model.add(Activation('relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))
#layer11 2*2*512
model.add(Conv2D(512, (3, 3), padding='same',kernel_regularizer=regularizers.l2(weight_decay)))
model.add(Activation('relu'))
model.add(BatchNormalization())
model.add(Dropout(0.4))
#layer12 2*2*512
model.add(Conv2D(512, (3, 3), padding='same',kernel_regularizer=regularizers.l2(weight_decay)))
model.add(Activation('relu'))
model.add(BatchNormalization())
model.add(Dropout(0.4))
#layer13 2*2*512
model.add(Conv2D(512, (3, 3), padding='same',kernel_regularizer=regularizers.l2(weight_decay)))
model.add(Activation('relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.5))
#layer14 1*1*512
model.add(Flatten())
model.add(Dense(512, kernel_regularizer=regularizers.l2(weight_decay)))
model.add(Activation('relu'))
model.add(BatchNormalization())
#layer15 512
model.add(Dense(512, kernel_regularizer=regularizers.l2(weight_decay)))
model.add(Activation('relu'))
model.add(BatchNormalization())
#layer16 512
model.add(Dropout(0.5))
model.add(Dense(10))
model.add(Activation('softmax'))

sgd = SGD(lr=1e-3, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(loss='categorical_crossentropy', optimizer=sgd,metrics=['accuracy'])
model.summary()
```

```python
history = model.fit(
    x_train[index_train], y_train[index_train], 
    epochs=nb_epoch, batch_size=batch_size, 
    validation_data=(x_test[index_test], y_test[index_test]), 
    verbose=1)
```

```python
model_name = 'VGG16_baseline_norm'
model_note = "SGD(lr=1e-3, decay=1e-6, momentum=0.9, nesterov=True)"
model_hist = f"training_log/history_{model_name}_" + "0607"

model.save(f'{model_name}.h5')
history.history["notes"] = model_note
with open(model_hist, "wb") as fo:
    pickle.dump(history.history, fo)
```

```python
history_files = [
    "training_log/history_VGG16_baseline_norm_1e-2_0606", 
    "training_log/history_VGG16_baseline_norm_1e-2_nodecay_0606", 
    "training_log/history_VGG16_baseline_norm_1e-3_0606", 
    "training_log/history_VGG16_baseline_norm_1e-3_nodecay_0606", 
    "training_log/history_VGG16_baseline_norm_1e-4_0606", 
    "training_log/history_VGG16_baseline_norm_1e-4_nodecay_0606", 
]
x_axis = np.arange(50) + 1
colors = ["#6FAB9C", "#E36A7A", "#5973DF"]
labels = [
    "Learning rate=1e-2 decay=1e-6",
    "Learning rate=1e-2",
    "Learning rate=1e-3 decay=1e-6",
    "Learning rate=1e-3",
    "Learning rate=1e-4 decay=1e-6",
    "Learning rate=1e-4",
]
plt.figure(figsize=(30, 10), dpi=100)
plt.subplot(131)
plt.xlabel("(a) Training accuracy")
for i, history_file in enumerate(history_files):
    with open(history_file, "rb") as foo:
        history = pickle.load(foo)
    plt.xlim([1, 50])
    plt.ylim([0.1, 1.0])
    line_style = "--" if "nodecay" in history_file else "-"
    plt.plot(x_axis, history['accuracy'], label=labels[i], linestyle=line_style, c=colors[i//2])
plt.grid()
plt.legend(loc='lower right')
plt.subplot(132)
plt.xlabel("(b) Testing accuracy")
for i, history_file in enumerate(history_files):
    with open(history_file, "rb") as foo:
        history = pickle.load(foo)
    plt.xlim([1, 50])
    plt.ylim([0.1, 0.9])
    line_style = "--" if "nodecay" in history_file else "-"
    plt.plot(x_axis, history['val_accuracy'], label=labels[i], linestyle=line_style, c=colors[i//2])
plt.grid()
plt.legend(loc='lower right')
plt.subplot(133)
plt.xlabel("(c) Training loss")
for i, history_file in enumerate(history_files):
    with open(history_file, "rb") as foo:
        history = pickle.load(foo)
    plt.xlim([1, 50])
    plt.ylim([1, 9])
    line_style = "--" if "nodecay" in history_file else "-"
    plt.plot(x_axis, history['loss'], label=labels[i], linestyle=line_style, c=colors[i//2])
plt.grid()
plt.legend(loc='upper right')
plt.subplots_adjust(left=0.02, bottom=0.05, right=0.99, top=0.99, wspace=0.1, hspace=0.1)
plt.savefig("learning_rate.png", dpi=200)
plt.show()
```
