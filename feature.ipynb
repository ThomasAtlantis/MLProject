```python
import os
os.environ['CUDA_VISIBLE_DEVICES'] = '1'

import tensorflow, platform
print(f"{platform.version()}, Python {platform.python_version()}, TensorFlow {tensorflow.__version__}")

import numpy as np
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.models import load_model, Model
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt

from cuml.decomposition import PCA
from cuml.manifold import TSNE
from cuml.cluster import KMeans
```

```python
model = load_model("VGG16_baseline_norm.h5")
```

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
x_train = x_train.astype('float32') / 255.0
y_train = to_categorical(y_train, 10)
x_test = x_test.astype('float32') / 255.0
y_test = to_categorical(y_test, 10)
```

```python
score = model.evaluate(x_test, y_test, verbose=1)
print('Test score:', score[0])
print('Test accuracy:', score[1])
```

```python
model.summary()
```

```python
layers = [
    "max_pooling2d_10", #  2 out
    "max_pooling2d_11", #  4 out
    "max_pooling2d_12", #  7 out
    "max_pooling2d_13", # 10 out
    "max_pooling2d_14", # 13 out
    "dense_7",          # 15 out
]
```

```python
indexes = np.random.choice(np.arange(x_test.shape[0]), size=1000, replace=False)
labels = np.argmax(y_test[indexes], axis=-1)
for layer in layers:
    _model = Model(inputs=model.input, outputs=model.get_layer(layer).output)
    result = _model.predict(x_test[indexes])
    result = np.reshape(result, (1000, -1))
    result = PCA(n_components=32).fit_transform(result)
    result = TSNE(n_components=2).fit_transform(result)
    np.save(f"feature_maps/{layer}.layer", result)
    plt.scatter(result[:, 0], result[:, 1], c=np.argmax(y_test[indexes], axis=-1))
    plt.show()
```

```python
plt.figure(figsize=(20, 10), dpi=80)
for i, layer in enumerate(layers):
    result = np.load(f"feature_maps/{layer}.layer.npy")
    plt.subplot(2, 3, i + 1)
    plt.scatter(result[:, 0], result[:, 1], c=labels, s=40, marker='.')
    plt.xticks([])
    plt.yticks([])
    plt.xlabel(f"({'abcdef'[i]})", fontsize=16)
plt.subplots_adjust(left=0.01, bottom=0.05, right=0.99, top=0.99, wspace=0.1, hspace=0.1)
plt.savefig("tsne.png", dpi=200)
plt.show()
```
